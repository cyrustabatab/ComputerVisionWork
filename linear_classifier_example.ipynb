```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = ['dog','cat','panda']
np.random.seed(1)
```

```python
W = np.random.randn(3,3072)
b = np.random.randn(3)
```

```python
orig = cv2.imread('beagle.png')
image = cv2.resize(orig,(32,32)).flatten()
```

```python
scores = W.dot(image) + b
```

```python
for label,score in zip(labels,scores):
    print(f"[INFO] {label}: {score:.2f}")
```

```python
orig= cv2.cvtColor(orig,cv2.COLOR_BGR2RGB)
```

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.imshow(orig)
ax.text(20,20,f"Label: {labels[np.argmax(scores)]}",fontsize=30,color='lightgreen')
ax.axis('off')
```

# Multi-class SVM Loss

$$ f(x_i,W,b) = Wx_i + b $$

$$ L_i = \sum_{j != y_i}max\big(0,s_j - s_{y_i} + 1\big) $$

$$ L = \frac{1}{N}\sum_{i=1}^N L_i $$

# Cross-entropy Loss and Softmax Classifier

$$ f(x_i,W) = Wx_i $$

$$ L_i = -log\big(e^{s_{y_i}}/\sum_je^{s_j}\big) $$
